for draw the this formula's chart : f(x,y) = (x+y) / (x^2 + y^2 + 1) in 3d's chart
we have to call two modules from python:
    1. 'matplotlib' module for draw the chart
    2. 'numpy' module for working with range of numbers and matrixes 

```python
from matplotlib import pyplot as plt
import numpy as np
```

It can be understood that our range for x and y vector is from -5 to +5 by the given chart.
so we can use the 'linespace' method from 'numpy' module for make a array of range from -5 to +5 with any number of point that we want.


    linespace(start, end, number)

let x have n point and y have m point.
for draw the 3d chart we must to know that in any point in x vector , we have m points in y vector and that repeat for all points of x
and for any point of y we have n point from x and it's true for all points of y.
that means we have two matrix mxn that one of them is for y and another is for x.
for create that matrix we can use the 'meshgrid' method from 'numpy' module that do this work for us.

```python
x = np.linspace(-5,5,100)
# print(x)
y = np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
print(X)
```
